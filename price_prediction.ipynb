```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
raw_zillow_data = pd.read_csv('ZHVI All Homes Zillow.csv')
# Assuming the dataset has columns 'feature1', 'feature2', ..., 'target'
raw_zillow_data
```

```python
raw_zillow_data = raw_zillow_data.dropna()  # Drop rows with missing values
# X = raw_zillow_data.drop(columns=['target'])  # Features
raw_zillow_data
```

```python
def extract_year(col):
    if isinstance(col, str) and len(col) == 10 and col[:4].isdigit() and col[4] == '-':
        return col[:7]
    return col

raw_zillow_data.columns = [extract_year(col) for col in raw_zillow_data.columns]
raw_zillow_data
```

```python
import datetime

current_year = datetime.datetime.now().year
years_to_keep = [str(year) for year in range(current_year - 5, current_year + 1)]  # e.g., ['2023', '2024', '2025']

# Identify columns in yyyy-mm format and in the desired years
cols_to_keep = [
    col for col in raw_zillow_data.columns
    if (len(col) == 7 and col[:4] in years_to_keep and col[4] == '-')
    or col in ['RegionName', 'StateName']  # keep identifier columns
]

filtered_data = raw_zillow_data[cols_to_keep]
filtered_data
```

```python
import plotly.graph_objs as go
from prophet import Prophet

future_months = 60  # Number of months to forecast

# 1. Calculate state-level monthly averages
value_columns = [col for col in filtered_data.columns if col not in ['RegionName', 'StateName']]
state_monthly_avg = filtered_data.groupby('StateName')[value_columns].mean()

fig = go.Figure()

for state in state_monthly_avg.index:
    ts = state_monthly_avg.loc[state].dropna()
    if len(ts) < 2:
        continue

    prophet_data = pd.DataFrame({
        'ds': ts.index,
        'y': ts.values.astype(float)
    })
    prophet_data['ds'] = pd.to_datetime(prophet_data['ds'], format='%Y-%m')

    model = Prophet()
    model.fit(prophet_data)

    future = model.make_future_dataframe(periods=future_months, freq='M')
    forecast = model.predict(future)

    fig.add_trace(go.Scatter(
        x=forecast['ds'],
        y=forecast['yhat'],
        mode='lines',
        name=state,
        hovertemplate='State: %{text}<br>Date: %{x|%Y-%m}<br>Price: $%{y:,.0f}',
        text=[state]*len(forecast)
    ))

fig.update_layout(
    title='State-Level Average House Price Forecasts (Interactive)',
    xaxis_title='Date',
    yaxis_title='Predicted Average Price',
    hovermode='x unified',
    legend_title='State',
    height=700
)
fig.update_yaxes(tickprefix="$", separatethousands=True)
fig.show()
```
